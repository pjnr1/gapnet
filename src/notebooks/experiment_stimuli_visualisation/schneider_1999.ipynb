```python
import matplotlib.pyplot as plt

from data_creation.experiments.schneider_1999 import generate_stimulus, MARKER_DURATIONS
from data_creation.time.time import generate_time_vector

fs = 100e3
gap_length = 6.0e-3
marker_duration = MARKER_DURATIONS[-2]
t = generate_time_vector(0.0, 1.0, fs)

gap, gate_gap = generate_stimulus(t,
                        marker_duration=marker_duration,
                        gap_length=gap_length)
no_gap, gate_no_gap = generate_stimulus(t,
                           marker_duration=marker_duration,
                           gap_length=0.0e-3)

fig, axs = plt.subplots(2, 1)
axs[0].plot(t, gap)
axs[0].plot(t, gate_gap)
axs[1].plot(t, no_gap)
axs[1].plot(t, gate_no_gap)
for ax in axs:
    ax.set_xlim([0.19, 0.61])
plt.show()
```
